# Tratamiento de nulos con pandas 

En pandas podemos trabajar con los nulos, los cuales son valores faltantes o información perdida.

```python
semana = pd.date_range(start = "2024-04-08", end = "2024-04-14")

df_resumen_ventas = pd.DataFrame(
    data = {
        "clientes": [10, np.nan, 12, 16, 20, np.nan, 14],
        "pedidos": [4, 2, 14, 10, 1, 10, 12],
        "ingresos": [np.nan, 20000, 4500, 7000, np.nan, 5900, 12000]
    },
    index=semana)

df_resumen_ventas
    
```

```python
#Vemos que valores son nulos
df_resumen_ventas.isnull()
```

```python
#Cálculo de cuántas filas contienen nulos
df_resumen_ventas.isnull().sum()
```

```python
#Deja solo las filas sin nulos
df_resumen_ventas.dropna()
```

```python
#Deja solo las columnas sin nulos
df_resumen_ventas.dropna(axis=1)
```

```python
#Rellenar los valores nulos con 0
df_resumen_ventas.fillna(0)
```

```python
df_resumen_ventas.median()
```

```python
#Reemplaza la media con la media de cada columna
df_resumen_ventas.fillna(df_resumen_ventas.median())
```

```python
#Se puede interpolar, aunque solo es recomendable cuando matemáticamente la operación tiene sentido
df_resumen_ventas.interpolate() #Obviamente si la primera fila tiene nulo no cambiará ya que no tiene interpelación posible 
```

```python
#Veremos como actuar cuando tenemos columnas cualitativas (categorías) y cuantitativas (numéricas)
dict_articulos = {
    "Producto": ["A", "B", "C", "D", "E"],
    "Cantidad Vendida": [100, np.nan, 150, np.nan, 200],
    "Precio Unitario": [10.5, 15.2, 12.7, np.nan, 9.8],
    "Categoría": ["Electrónica", "Ropa", np.nan, "Electrónica", "Ropa"]
}

df_articulos = pd.DataFrame(dict_articulos)

df_articulos
```

```python
#Calculamos el promedio de la cantidad vendida, excluyendo los valores nulos
promedio_cantidad_vendida = df_articulos["Cantidad Vendida"].mean()
promedio_cantidad_vendida
```

```python
#Ahora reemplazaremos los valores nulos en "Cantidad Vendida" con cero y los de "Precio Unitario" con la media de esa columna
df_articulos["Cantidad Vendida"] = df_articulos["Cantidad Vendida"].fillna(0)
precio_unitario_promedio = df_articulos["Precio Unitario"].mean()
df_articulos["Precio Unitario"] = df_articulos["Precio Unitario"].fillna(precio_unitario_promedio)

#Para las columnas categóricas podríamos reemplazar los valores nulos con una categoróa especial como "Desconocido"
df_articulos["Categoría"] = df_articulos["Categoría"].fillna("Desconocido")

df_articulos
```

```python
#Calculamos el promedio de la cantidad vendida después de reemplazar los valores nulos
promedio_cantidad_vendida = df_articulos["Cantidad Vendida"].mean()
promedio_cantidad_vendida
```

### Ejercicio 1.5

Con pandas se puede trabajar con archivos JSON lectura y escritura.

JSON es un formato de texto que forma parte del sistema de JavaScript y que se deriva de su sintaxis, pero no tiene como objetivo la creación de programas, sino el acceso, almacenamiento e intercambio de datos. 

Normalmente la información que se guarda en un archivo JSON no es de grandes dimensiones. Una práctica muy común es que las filas de datos de una tabla / dataframe sean transformadas a un JSON o leidas desde uno.

```python
#Leamos un csv muy grande, extraigamos un sub dataset más pequeño y creemos un JSON por cada fila
df_ecommerce_sale_report = pd.read_csv("https://raw.githubusercontent.com/Andresmup/ArchivosDataScience/main/ecommerce_sale_report.csv", index_col="index", low_memory=False)
df_ecommerce_sale_report.shape #Veamos la cantidad de filas y columnas
```

```python
df_ecommerce_sample = df_ecommerce_sale_report.sample(n = 5, random_state = 42)
df_ecommerce_sample
```

```python
#Crear la carpeta "data_parquet" si no existe
if not os.path.exists("data_jason"):
    os.makedirs("data_jason")
```

```python
#Con un ciclo for veamos e iteremos sobre los 5 elementos de df_ecommerce_sample

for index, row in df_ecommerce_sample.iterrows():
    data = {} #Cada JSON empieza como un diccionario vacío 
    for column, value in row.items(): #Hay que convertir las filas en elementos para rellenar el JSON

     #Esta parte es para evitar un posible error ya que los JSON son muy sensibles a los nulos
        if pd.isna(value): #Convertir NaN a None
            value = None
        elif column == "ship-posta-code": #Convertir ship-postal-code a cadena
            value = str(value)
        data[column] = value

    #Convertir cada diccionario data a una cadena JSON
    json_data = json.dumps(data)

    #Guardar la cadena JSON en un archivo
    json_filename = f"data_jason/fila_{index}.json"
    with open(json_filename, "w") as json_file:
        json_file.write(json_data)

    print(data)
```

### Ejercicio 1.6

Un archivo muy utilizado cuando se tratan archivos grandes es el .parquet. Normalmente este formato suele ser el preferido de cara a almacenar información en lagos de datos.

```python
#Leemos el archivo
df_ecommerce_sale_report = pd.read_csv("https://raw.githubusercontent.com/Andresmup/ArchivosDataScience/main/ecommerce_sale_report.csv", index_col="index", low_memory=False)
df_ecommerce_sale_report.shape #Veamos la cantidad de filas y columnas
```

```python
#Crear la carpeta "data_parquet" si no existe 
if not os.path.exists("data_parquet"):
    os.makedirs("data_parquet")
```

```python
#Tamaño de la partición (número de filas por archivo) 
tamano_particion = 10000

#Calcular el número total de particiones necesarias
num_particiones = len(df_ecommerce_sale_report) // tamano_particion + 1

#Particionar y guardar cada partición como un archivo Parquet
for i in range(num_particiones):
    inicio = i * tamano_particion
    fin = min((i + 1) * tamano_particion, len(df_ecommerce_sale_report))
    particion = df_ecommerce_sale_report.iloc[inicio:fin]
    particion.to_parquet(f"data_parquet/particion_{i}.parquet")
```
